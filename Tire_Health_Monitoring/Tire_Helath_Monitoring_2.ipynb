```python
file_list = os.listdir()
file_list
```

```python
ls = [0 for i in range(len(file_list))]
x = 0
for file_path in file_list:
    df = pd.concat([pd.read_csv(file_path)])
    #df['Name'] = file_path
    ls[x] = df
    x = x+1
```

```python
df = pd.concat(ls)
df = df.reset_index(drop = True)
df.head()
```

```python
result_df = df[~(df['Velocity (Km/h)']>=80) & ~(df['Velocity (Km/h)']==0)]
```

```python
result_df
```

```python
result_df = result_df.groupby('Time (-)').mean().dropna()
```

```python
result_df.shape
```

```python
result_df.reset_index(inplace=True)
```

```python
result_df
```

```python
result_df.boxplot(column = ['Velocity (Km/h)'],vert=0)
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(result_df,test_size=.3,random_state = 42)
```

```python
# train_y = train['Velocity (Km/h)']
# train_x = train[['X (G)','Y (G)','Z (G)']]
# test_x = test[['X (G)','Y (G)','Z (G)']]
# test_y = test['Velocity (Km/h)']
```

```python
train_x = train.drop(['Velocity (Km/h)','Time (-)','Distance (m)'],axis=1)
train_y = train['Velocity (Km/h)']
test_x = test.drop(['Velocity (Km/h)','Time (-)','Distance (m)'],axis=1)
test_y = test['Velocity (Km/h)']
```

```python
train.columns.drop(['Time (-)','Velocity (Km/h)','Distance (m)'])
```

```python
from matplotlib import pyplot as plt
from sklearn.svm import SVR

def f_importances(coef, names):
    imp = coef
    imp,names = zip(*sorted(zip(imp,names)))
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
    plt.show()

features_names = train.columns.drop(['Time (-)','Velocity (Km/h)','Distance (m)'])
svm = SVR(kernel='linear')
svm.fit(train_x,train_y)
```

```python
plt.figure(figsize=(20,15))
f_importances(svm.coef_[0],features_names)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(train_x,train_y)
```

```python
plt.figure(figsize=(20,15))
feature_imp = rf.feature_importances_
def f_importances(coef, names):
    imp = coef
    imp,names = zip(*sorted(zip(imp,names)))
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
f_importances(feature_imp,features_names)
plt.title('RandomForest Feature Importance',fontsize=20)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test_y,rf.predict(test_x)))
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(random_state=42)
xgb.fit(train_x,train_y)
```

```python
mean_squared_error(test_y,xgb.predict(test_x))
```

```python
plt.figure(figsize=(20,15))
feature_imp = xgb.feature_importances_
f_importances(feature_imp,features_names)
plt.title('XGBoost Feature Importance',fontsize=20)
plt.show()
```

```python
from lightgbm import LGBMRegressor
lgb = LGBMRegressor()
lgb.fit(train_x,train_y)
```

```python
plt.figure(figsize=(20,15))
feature_imp = lgb.feature_importances_
f_importances(feature_imp,features_names)
plt.title('LGBM Feature Importance',fontsize=20)
plt.show()
```
