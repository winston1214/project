```python
def show_height(ax,fontsize=10):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=fontsize)

import os
import matplotlib.font_manager as fm
from matplotlib import rc

def set_font(font_file='malgun.ttf'): #  
    font_dir = 'c:/Windows/Fonts/'
    font_name = fm.FontProperties(fname=os.path.join(font_dir, font_file)).get_name()
    rc('font', family=font_name)
set_font()
```

# Preprocessing

```python
dataset['family_size'] = dataset['sibsp'] + dataset['parch'] +1
```

```python
dataset['family_size'].describe()
```

```python
# 1 : isalone , 2~4 : small, 5~ big
dataset['family_size'] = pd.cut(dataset['family_size'],[0,1,5,12],labels=['isalone','small','big'])
```

```python
# age_dic = {'10s': 0 ,'20s':1,'30s':2 , '40s':3,'50s':4,'old':5}
# dataset['age'] = dataset['age'].map(age_dic)
```

```python
title_map={ 'capt':'officer',
           'col':'officer'
           ,'major':'officer'
           ,'dr':'officer'
           ,'rev':'officer'
           ,'sir':'officer',
           'jonkheer': 'royalty',
          'the countess':'royalty',
          'dona': 'royalty',
          'lady': 'royalty',
          'don': 'royalty',
          'mr': 'mr',
          'mrs':'mrs',
          'ms':'mrs',
          'mme':'mrs',
          'miss':'miss',
          'mlle':'miss',
          'master':'master'}
```

```python
dataset['name'] = dataset['name'].str.lower()
```

```python
dataset['name'] = dataset['name'].map(title_map)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python
df = dataset[['age_range','name','family_size','sibsp','parch']]
```

```python
df = pd.concat([df,pd.get_dummies(df.drop(['age_range','sibsp','parch'],axis=1))],axis=1)
df.drop(['name','family_size'],axis=1,inplace=True)
```

```python
idx = df[df['age_range'].isnull()].index
```

```python
train_x = df.drop('age_range',axis=1)[df['age_range'].notnull()]
```

```python
train_y = df[df['age_range'].notnull()]['age_range']
```

```python
test_x = df.drop('age_range',axis=1)[df['age_range'].isnull()]
test_y = df[df['age_range'].isnull()]['age_range']
```

```python
rf = RandomForestClassifier(n_estimators=400,max_depth=20)
xgb = XGBClassifier(n_estimators=400,max_depth=20)
```

```python
train_x2,valid_x,train_y2,valid_y = train_test_split(train_x,train_y,test_size=.2,random_state=42)
```

```python
rf.fit(train_x2,train_y2)
rf_pred = rf.predict(valid_x)
xgb.fit(train_x2,train_y2)
xgb_pred = xgb.predict(valid_x)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(valid_y.values,rf_pred))
print(accuracy_score(valid_y.values,xgb_pred))
```
